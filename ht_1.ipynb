```python
import pandas as pd
```

Ниже представлен минимальный список колонок, которые должны находиться в итоговом датасете после склейки и агрегации данных. По своему усмотрению вы можете добавить дополнительные к этим колонки.

- AGREEMENT_RK — уникальный идентификатор объекта в выборке;
- TARGET — целевая переменная: отклик на маркетинговую кампанию (1 — отклик был зарегистрирован, 0 — отклика не было);
- AGE — возраст клиента;
- SOCSTATUS_WORK_FL — социальный статус клиента относительно работы (1 — работает, 0 — не работает);
- SOCSTATUS_PENS_FL — социальный статус клиента относительно пенсии (1 — пенсионер, 0 — не пенсионер);
- GENDER — пол клиента (1 — мужчина, 0 — женщина);
- CHILD_TOTAL — количество детей клиента;
- DEPENDANTS — количество иждивенцев клиента;
- PERSONAL_INCOME — личный доход клиента (в рублях);
- LOAN_NUM_TOTAL — количество ссуд клиента;
- LOAN_NUM_CLOSED — количество погашенных ссуд клиента.

Будьте внимательны при сборке датасета: это реальные банковские данные, в которых могут наблюдаться дубли, некорректно заполненные значения или значения, противоречащие друг другу. Для получения качественной модели необходимо предварительно очистить датасет от такой информации.

```python
df_clients = pd.read_csv('data/D_clients.csv')
df_close_loan = pd.read_csv('data/D_close_loan.csv')
df_job = pd.read_csv('data/D_job.csv')
df_last_credit = pd.read_csv('data/D_last_credit.csv')
df_loan = pd.read_csv('data/D_loan.csv')
df_pens = pd.read_csv('data/D_pens.csv')
df_salary = pd.read_csv('data/D_salary.csv')
df_work = pd.read_csv('data/D_work.csv')
df_target = pd.read_csv('data/D_target.csv')
df_clients.describe(include='object')
```

```python
df_merged = (pd.merge(df_clients, df_target, left_on='ID', right_on='ID_CLIENT', how='inner')
.merge(df_salary, on='ID_CLIENT', how='left')
.merge(df_loan, on='ID_CLIENT', how='left')
.merge(df_close_loan, on='ID_LOAN', how='left')
.merge(df_job, on='ID_CLIENT', how='left')
.merge(df_salary, on='ID_CLIENT', how='left')
.drop([ 'ID_CLIENT', 'MARITAL_STATUS', 'REG_ADDRESS_PROVINCE', 
                'FACT_ADDRESS_PROVINCE', 'POSTAL_ADDRESS_PROVINCE', 
                'GEN_INDUSTRY',
                'GEN_TITLE', 'JOB_DIR', 'ID_LOAN',
                'FAMILY_INCOME_y','PERSONAL_INCOME_y'], axis=1))

#df_merged.describe(include='object')
df_merged.describe() # видно, что "ушло" часть признаков где 
duplicate_count = df_merged['ID'].isna().sum()
duplicate_count

#unique_values = df_merged['EDUCATION'].unique()
#df_merged['EDUCATION'] = pd.Categorical(df_merged['EDUCATION']).codes
#df_merged['FAMILY_INCOME'] = pd.Categorical(df_merged['FAMILY_INCOME_x']).codes

#df_merged.drop('FAMILY_INCOME_x', axis=1, inplace=True)


df_merged.describe()
```

```python

df_merged['LOAN_NUM_TOTAL'] = df_merged.groupby('ID')['ID'].transform('count')
df_merged['LOAN_NUM_CLOSED'] = df_merged.groupby('ID')['CLOSED_FL'].transform(lambda x: (x == 1).sum())
df_merged = df_merged.drop_duplicates(subset='ID')

```

```python
df_merged.rename(columns={'PERSONAL_INCOME_x': 'PERSONAL_INCOME'}, inplace=True)
df_merged.drop(columns = ["AGREEMENT_RK", "ID"], inplace=True)
df_merged
```

```python
df_merged.describe()
```

```python
# чистим значения в колонке WorkTime и заполняем средними
# PS не нужно работать больше 360 часов в месяц
df_merged.loc[df_merged['WORK_TIME'] > 360, 'WORK_TIME'] = 360

mean_value = df_merged['WORK_TIME'].mean()

df_merged['WORK_TIME'].fillna(mean_value, inplace=True)
df_merged.describe()
```

```python
df_merged.to_csv("dataset.csv", index=False)
```
